# ***data model TF & IDF( Test Frequency & Inverse Document Frequency)***

# *data preparation()*
Load your dataset

Combine relevant features (genres + ratings)

Handle missing values

```python
import pandas as pd

# Load dataset
df = pd.read_csv("netflix_titles.csv")

# Combine genre and rating into a single text column
df['combined_features'] = df['listed_in'] + " " + df['rating']

# Fill any missing values with an empty string
df['combined_features'] = df['combined_features'].fillna("")
```

# Convert Text Data into TF-IDF Vectors()
Use TfidfVectorizer to transform text into numerical data.

Stop words (common words like "the", "and", etc.) are removed.

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Convert text into numerical vectors using TF-IDF
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(df['combined_features'])

# Print shape of the matrix (should be movies × unique words)
print("TF-IDF Matrix Shape:", tfidf_matrix.shape)
```

# Compute Cosine Similarity()
Cosine Similarity finds how similar two movies are.

A higher score means movies are more alike

```python
from sklearn.metrics.pairwise import cosine_similarity

# Compute cosine similarity between all movies
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Print similarity matrix shape
print("Cosine Similarity Matrix Shape:", cosine_sim.shape)
```

# Create the Recommendation Function()
Find the most similar movies based on TF-IDF & Cosine Similarity

```python
def recommend_movies(title, df, cosine_sim):
    try:
        # Get the index of the selected movie
        index = df[df['title'] == title].index[0]

        # Get similarity scores for all movies
        similarity_scores = list(enumerate(cosine_sim[index]))

        # Sort by similarity score (highest first, ignore first entry as it's the same movie)
        similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)[1:6]

        # Get the top 5 most similar movies
        movie_indices = [i[0] for i in similarity_scores]

        return df.iloc[movie_indices][['title', 'listed_in', 'rating']]

    except IndexError:
        return "Movie not found in dataset!"

# Example usage
print(recommend_movies("Stranger Things", df, cosine_sim))
```

# ***Hidden Gems Recommendation Code()***

```python
def find_hidden_gems(title, df, cosine_sim, min_popularity=10, k=5):
    try:
        # Get the index of the movie
        index = df[df['title'] == title].index[0]

        # Get similarity scores for all movies
        similarity_scores = list(enumerate(cosine_sim[index]))

        # Sort movies by similarity score (highest first)
        similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)[1:]

        # Get indices of all movies sorted by similarity
        movie_indices = [i[0] for i in similarity_scores]

        # Filter movies that are less known (hidden gems)
        hidden_gems = df.iloc[movie_indices]

        # Apply filtering criteria (low popularity, niche genres, etc.)
        hidden_gems = hidden_gems[hidden_gems['release_year'] > 2000]  # Example: Only newer movies
        hidden_gems = hidden_gems[hidden_gems['rating'].isin(['TV-MA', 'TV-14', 'R', 'PG-13'])]  # Example: Select specific ratings

        # Limit to top K hidden gems
        return hidden_gems[['title', 'listed_in', 'release_year']].head(k)

    except IndexError:
        return "Movie not found!"

# Example usage:
print("Hidden Gems similar to 'Dark':")
print(find_hidden_gems("Dark", df, cosine_sim, k=5))
```

# movie recommendations by using different attributes()

```python
def find_hidden_gems(df, cosine_sim, type_=None, title=None, cast=None, director=None,
                     release_year=None, rating=None, duration=None, hashtags=None, description=None, k=5):
    # Filter based on user inputs
    filtered_df = df.copy()

    if type_:
        filtered_df = filtered_df[filtered_df['type'].str.contains(type_, case=False, na=False)]
    if title:
        filtered_df = filtered_df[filtered_df['title'].str.contains(title, case=False, na=False)]
    if cast:
        filtered_df = filtered_df[df['cast'].str.contains(cast, case=False, na=False)]
    if director:
        filtered_df = filtered_df[df['director'].str.contains(director, case=False, na=False)]
    if release_year:
        filtered_df = filtered_df[filtered_df['release_year'] == release_year]
    if rating:
        filtered_df = filtered_df[filtered_df['rating'] == rating]
    if duration:
        filtered_df = filtered_df[filtered_df['duration'].str.contains(duration, case=False, na=False)]
    if hashtags:
        filtered_df = filtered_df[filtered_df['listed_in'].str.contains(hashtags, case=False, na=False)]
    if description:
        filtered_df = filtered_df[filtered_df['description'].str.contains(description, case=False, na=False)]

    # Get indices of filtered movies
    movie_indices = filtered_df.index.tolist()

    # Get similarity scores for filtered movies
    similarity_scores = []
    for idx in movie_indices:
        similarity_scores.extend(list(enumerate(cosine_sim[idx])))

    # Sort by similarity score (descending order)
    similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)

    # Get unique movie indices
    unique_movie_indices = []
    for score in similarity_scores:
        if score[0] not in unique_movie_indices:
            unique_movie_indices.append(score[0])
        if len(unique_movie_indices) >= k:
            break

    # Return top K hidden gems
    hidden_gems = df.iloc[unique_movie_indices]
    return hidden_gems[['title', 'listed_in', 'release_year', 'rating']]

# Example usage:
print("Hidden Gems based on filters:")
print(find_hidden_gems(df, cosine_sim, type_="Movie", director="Christopher Nolan", k=5))
```

```python
import pandas as pd

def preprocess_data(file_path):
    # Load dataset
    df = pd.read_csv(file_path)

    # Remove unnecessary columns
    columns_to_remove = ['show_id', 'date_added', 'IMDb score']
    df = df.drop(columns=[col for col in columns_to_remove if col in df.columns], errors='ignore')

    # Drop rows with missing values
    df = df.dropna()

    return df

def filter_by_genre(df, genre):
    # Filter dataset by genre
    return df[df['listed_in'].str.contains(genre, case=False, na=False)]

def filter_by_rating(df, rating):
    # Filter dataset by rating
    return df[df['rating'].str.contains(rating, case=False, na=False)]

def filter_by_hashtag(df, keyword):
    # Filter dataset by cast or director
    return df[df[['cast', 'director']].apply(lambda x: x.str.contains(keyword, case=False, na=False)).any(axis=1)]

def detect_hidden_gems(df):
    # Identify hidden gems based on rarity factors
    rare_countries = ['Norway', 'Denmark', 'Finland', 'Iceland', 'New Zealand', 'South Korea', 'Argentina', 'Turkey', 'Greece', 'Portugal']
    df_hidden = df[df['country'].isin(rare_countries)]

    # Consider filtering by older but well-rated content
    df_hidden = df_hidden[df_hidden['release_year'] < 2020]  # Adjusted threshold to allow more results

    # Handle cases where country info is missing
    if df_hidden.empty:
        df_hidden = df[df['release_year'] < 2020]  # Fallback to showing older movies from any country

    return df_hidden

# Example usage
if __name__ == "__main__":
    file_path = "netflix_titles.csv"  # Updated dataset file name
    processed_df = preprocess_data(file_path)

    # Example: Filter by a specific genre
    genre = "thriller"
    filtered_by_genre_df = filter_by_genre(processed_df, genre)

    # Example: Filter by a specific rating
    rating = "PG-13"
    filtered_by_rating_df = filter_by_rating(filtered_by_genre_df, rating)

    # Example: Filter by a specific cast or director
    hashtag = "Christopher Nolan"
    filtered_by_hashtag_df = filter_by_hashtag(filtered_by_rating_df, hashtag)

    # Identify hidden gems
    hidden_gems_df = detect_hidden_gems(filtered_by_hashtag_df)

    print(hidden_gems_df.head())
```
